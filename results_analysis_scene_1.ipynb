```python
from wrapper_panda import PandaWrapper
from scenes import Scene

robot_wrapper = PandaWrapper()
rmodel, cmodel, vmodel = robot_wrapper()
scene = Scene()
cmodel = scene.create_scene(cmodel, 1)
rdata = rmodel.createData()
cdata = cmodel.createData()

def compute_end_effector_position_and_velocity(x):
    pin.forwardKinematics(rmodel, rdata, np.array(x[:7]), np.array(x[7:]))
    pin.updateFramePlacements(rmodel, rdata)
    # pin.framesForwardKinematics(rmodel, rdata, np.array(x[:7]), np.array(x[7:]))
    pin.updateGeometryPlacements(rmodel, rdata, cmodel, cdata, np.array(x[:7]))
    
    pose = cdata.oMg[cmodel.getGeometryId("panda2_leftfinger_0")].translation.copy()
    vel = pin.getFrameVelocity(rmodel, rdata, cmodel.geometryObjects[cmodel.getGeometryId("panda2_leftfinger_0")].parentFrame, pin.LOCAL_WORLD_ALIGNED).linear.copy()
    
    return pose, vel

endeff_pose_dist = []
endeff_vel_dist = []
endeff_pose_nocol = []
endeff_vel_nocol = []
endeff_pose_vel = []
endeff_vel_vel = []
for it in range(data_distance["T"]):
    pose, vel = compute_end_effector_position_and_velocity(data_distance["xs"][it])
    endeff_pose_dist.append(pose)
    endeff_vel_dist.append(vel)
    pose, vel = compute_end_effector_position_and_velocity(data_velocity["xs"][it])
    endeff_pose_vel.append(pose)
    endeff_vel_vel.append(vel)
    pose, vel = compute_end_effector_position_and_velocity(data_nocol["xs"][it])
    endeff_pose_nocol.append(pose)
    endeff_vel_nocol.append(vel)


endeff_pose_dist_x = [pose[0] for pose in endeff_pose_dist]
endeff_pose_dist_y = [pose[1] for pose in endeff_pose_dist]
endeff_pose_dist_z = [pose[2] for pose in endeff_pose_dist]

endeff_vel_dist_x = [pose[0] for pose in endeff_vel_dist]
endeff_vel_dist_y = [pose[1] for pose in endeff_vel_dist]
endeff_vel_dist_z = [pose[2] for pose in endeff_vel_dist]

endeff_pose_nocol_x = [pose[0] for pose in endeff_pose_nocol]
endeff_pose_nocol_y = [pose[1] for pose in endeff_pose_nocol]
endeff_pose_nocol_z = [pose[2] for pose in endeff_pose_nocol]

endeff_pose_vel_x = [pose[0] for pose in endeff_pose_vel]
endeff_pose_vel_y = [pose[1] for pose in endeff_pose_vel]
endeff_pose_vel_z = [pose[2] for pose in endeff_pose_vel]

endeff_vel_vel_x = [vel[0] for vel in endeff_vel_vel]
endeff_vel_vel_y = [vel[1] for vel in endeff_vel_vel]
endeff_vel_vel_z = [vel[2] for vel in endeff_vel_vel]

endeff_vel_nocol_x = [vel[0] for vel in endeff_vel_nocol]
endeff_vel_nocol_y = [vel[1] for vel in endeff_vel_nocol]
endeff_vel_nocol_z = [vel[2] for vel in endeff_vel_nocol]
```

```python
import hppfcl

collision_pairs = data_distance["collision_pairs"]
obstacle_id = collision_pairs[0][1]
link7_id = collision_pairs[0][0]
endeff_id = collision_pairs[1][0]
print(collision_pairs)
print(endeff_id)
```

```python
robot_wrapper = PandaWrapper()
rmodel, cmodel, vmodel = robot_wrapper()
scene = Scene()
cmodel = scene.create_scene(cmodel, 1)
rdata = rmodel.createData()
cdata = cmodel.createData()

def distance_to_obs(shape_id, q):
        
    pin.forwardKinematics(rmodel, rdata, np.array(q))
    pin.updateFramePlacements(rmodel, rdata)
    pin.updateGeometryPlacements(rmodel, rdata, cmodel, cdata)
    
        # Poses and geometries of the shapes
    shape_placement = cdata.oMg[shape_id]
    obs_placement = cdata.oMg[obstacle_id]
    
    req = hppfcl.DistanceRequest()
    res = hppfcl.DistanceResult()
    
    distance = hppfcl.distance(cmodel.geometryObjects[shape_id].geometry, shape_placement, cmodel.geometryObjects[obstacle_id].geometry, obs_placement, req, res)
    
    return distance    
    
```

```python
distance_obstacle_end_effector_dist = []
distance_obstacle_link7_dist = []
distance_obstacle_end_effector_vel = []
distance_obstacle_link7_vel = []

distance_obstacle_end_effector_nocol= []
distance_obstacle_link7_nocol = []

for it in range(data_distance["T"]):
    distance_obstacle_end_effector_dist.append(distance_to_obs(endeff_id, data_distance["xs"][it][:7]))
    distance_obstacle_link7_dist.append(distance_to_obs(link7_id, data_distance["xs"][it][:7]))
    
    distance_obstacle_end_effector_vel.append(distance_to_obs(endeff_id, data_velocity["xs"][it][:7]))
    distance_obstacle_link7_vel.append(distance_to_obs(link7_id, data_velocity["xs"][it][:7]))
    
    distance_obstacle_end_effector_nocol.append(distance_to_obs(endeff_id, data_nocol["xs"][it][:7]))
    distance_obstacle_link7_nocol.append(distance_to_obs(link7_id, data_nocol["xs"][it][:7]))
```

```python
# Plot the distance obstacles
fig = plt.figure(figsize=(10, 10))

plt.plot(distance_obstacle_end_effector_dist, '-' ,label='End Effector Distance, Distance Constraint')
plt.plot(distance_obstacle_link7_dist, '-' ,label='Link7 Distance, Distance Constraint')


plt.plot(distance_obstacle_end_effector_vel, '-' ,label='End Effector Distance, Velocity Constraint')
plt.plot(distance_obstacle_link7_vel, '-' ,label='Link7 Distance, Velocity Constraint')

plt.plot(distance_obstacle_end_effector_nocol, '-' ,label='End Effector Distance, No Constraint')
plt.plot(distance_obstacle_link7_nocol, '-' ,label='Link7 Distance, No Constraint')

plt.plot(np.zeros(data_distance["T"]), 'k--', label='0 distance')

# Add labels and title
plt.xlabel('Step')
plt.ylabel('Distance')
plt.title('Comparison of Distance Obstacles')
plt.legend()
```
