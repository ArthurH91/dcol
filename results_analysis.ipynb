```python
fig, axs = plt.subplots(7, 1, figsize=(10, 20))

for it, (key, val) in enumerate( x_dist.items()):
    axs[it].plot(val, label='data_distance')
    axs[it].plot(x_vel[key], label='data_velocity')
    axs[it].set_xlabel('Index')
    axs[it].set_ylabel('Value')
    axs[it].set_title(f'Comparison of the joint position of the distance and velocity constraints for joint {it}')

axs[0].legend()
plt.tight_layout()
plt.show()
```
