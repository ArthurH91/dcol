```python
import numpy as np
import matplotlib.pyplot as plt

# Function to create an ellipsoid
def plot_ellipsoid(ax, center, radii, rotation, resolution=50):
    u = np.linspace(0, 2 * np.pi, resolution)
    v = np.linspace(0, np.pi, resolution)
    x = radii[0] * np.outer(np.cos(u), np.sin(v))
    y = radii[1] * np.outer(np.sin(u), np.sin(v))
    z = radii[2] * np.outer(np.ones_like(u), np.cos(v))

    # Apply the rotation
    ellipsoid_points = np.array([x.flatten(), y.flatten(), z.flatten()])
    rotated_points = rotation @ ellipsoid_points

    # Reshape and translate the ellipsoid points to the center
    x_rot = rotated_points[0, :].reshape(x.shape) + center[0]
    y_rot = rotated_points[1, :].reshape(y.shape) + center[1]
    z_rot = rotated_points[2, :].reshape(z.shape) + center[2]

    # Plot the ellipsoid
    ax.plot_surface(x_rot, y_rot, z_rot, color='r', alpha=0.3, rstride=4, cstride=4)

# Ellipsoid parameters
DIM_Ellipsoids = data_distance["obstacle_dim"][0]  # radii of the ellipsoid
center = data_distance["obstacle_pose_translation"][0]  # position of the ellipsoid's center
rotation_matrix = data_distance["obstacle_pose_rotation"][0]  # identity matrix for no rotation

# Create the plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot end effector position for distance constraint
ax.plot(endeff_pose_dist_x, endeff_pose_dist_y, endeff_pose_dist_z, label='End Effector Pose Distance constraint')

# Plot end effector position for velocity constraint
ax.plot(endeff_pose_vel_x, endeff_pose_vel_y, endeff_pose_vel_z, label='End Effector Pose Velocity constraint')

# Plot end effector position for no collision constraint
ax.plot(endeff_pose_nocol_x, endeff_pose_nocol_y, endeff_pose_nocol_z, label='End Effector Pose No Collision constraint')

# Plot the ellipsoid
plot_ellipsoid(ax, center, DIM_Ellipsoids, rotation_matrix)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Comparison of End Effector Pose (Distance vs Velocity) with Ellipsoid')

ax.view_init(elev=10, azim=90)  # You can change these values to rotate the plot

ax.legend()
plt.show()
```

```python
def distance_to_obs(shape_id, q):
        
    pin.forwardKinematics(rmodel, rdata, np.array(q))
    pin.updateFramePlacements(rmodel, rdata)
    pin.updateGeometryPlacements(rmodel, rdata, cmodel, cdata)
    
        # Poses and geometries of the shapes
    shape_placement = cdata.oMg[shape_id]
    obs_placement = cdata.oMg[obstacle_id]
    
    req = hppfcl.DistanceRequest()
    res = hppfcl.DistanceResult()
    
    distance = hppfcl.distance(cmodel.geometryObjects[shape_id].geometry, shape_placement, cmodel.geometryObjects[obstacle_id].geometry, obs_placement, req, res)
    
    return distance    
    
```

```python
distance_obstacle_end_effector_dist = []
distance_obstacle_link7_dist = []
distance_obstacle_end_effector_vel = []
distance_obstacle_link7_vel = []

distance_obstacle_end_effector_nocol= []
distance_obstacle_link7_nocol = []

for it in range(data_distance["T"]):
    distance_obstacle_end_effector_dist.append(distance_to_obs(endeff_id, data_distance["xs"][it][:7]))
    distance_obstacle_link7_dist.append(distance_to_obs(link7_id, data_distance["xs"][it][:7]))
    
    distance_obstacle_end_effector_vel.append(distance_to_obs(endeff_id, data_velocity["xs"][it][:7]))
    distance_obstacle_link7_vel.append(distance_to_obs(link7_id, data_velocity["xs"][it][:7]))
    
    distance_obstacle_end_effector_nocol.append(distance_to_obs(endeff_id, data_nocol["xs"][it][:7]))
    distance_obstacle_link7_nocol.append(distance_to_obs(link7_id, data_nocol["xs"][it][:7]))
```
